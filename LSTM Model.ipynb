```python
# topographical charcterstics
#['gauge_lat' 'gauge_lon' 'elev_mean' 'slope_mean' 'area_gages2'
# 'area_geospa_fabric' 

gauge_lat = camels.read_constant_cols(
    gage_ids[:671], var_lst=['gauge_lat']
).ravel()
gauge_lon = camels.read_constant_cols(
    gage_ids[:671], var_lst=['gauge_lon']
).ravel()
elev_mean = camels.read_constant_cols(
    gage_ids[:671], var_lst=['elev_mean']
).ravel()
slope_mean = camels.read_constant_cols(
    gage_ids[:671], var_lst=['slope_mean']
).ravel()
area_geospa_fabric = camels.read_constant_cols(
    gage_ids[:671], var_lst=['area_geospa_fabric']
).ravel()
area_gages2 = camels.read_constant_cols(
    gage_ids[:671], var_lst=['area_gages2']
).ravel()

```

```python
# displaying heatmap
# geological Characterstics

geo_data = {
   'gauge_lat':gauge_lat,
   'gauge_lon': gauge_lon,
    'elev_mean': elev_mean,
    'slope_mean':slope_mean ,
    'area_gages2': area_gages2,
    'area_geospa_fabric':area_geospa_fabric

}
 
# form dataframe
geoData = pd.DataFrame(geo_data, columns=['gauge_lat', 'gauge_lon', 'elev_mean', 'slope_mean', 'area_gages2',
                                          'area_geospa_fabric'])
print("Dataframe is : ")
print(geoData)
 
# form correlation matrix
geo_matrix = geoData.corr()
print("Correlation matrix is : ")
print(geo_matrix)
dataplot = sb.heatmap(geo_matrix, cmap="YlGnBu", annot=True)
  
# displaying heatmap
mp.show()
```

```python
# climatic charcterstics
#'p_mean' 'pet_mean' 'p_seasonality' 'frac_snow'
# 'aridity' 'high_prec_freq' 'high_prec_dur' 'high_prec_timing'
 #'low_prec_freq' 'low_prec_dur' 'low_prec_timing',
    
p_mean = camels.read_constant_cols(
    gage_ids[:671], var_lst=['p_mean']
).ravel()
pet_mean = camels.read_constant_cols(
    gage_ids[:671], var_lst=['pet_mean']
).ravel()
p_seasonality = camels.read_constant_cols(
    gage_ids[:671], var_lst=['p_seasonality']
).ravel()
frac_snow = camels.read_constant_cols(
    gage_ids[:671], var_lst=['frac_snow']
).ravel()
aridity = camels.read_constant_cols(
    gage_ids[:671], var_lst=['aridity']
).ravel()
high_prec_freq = camels.read_constant_cols(
    gage_ids[:671], var_lst=['high_prec_freq']
).ravel()
high_prec_dur = camels.read_constant_cols(
    gage_ids[:671], var_lst=['high_prec_dur']
).ravel()
high_prec_timing = camels.read_constant_cols(
    gage_ids[:671], var_lst=['high_prec_timing']
).ravel()
low_prec_freq = camels.read_constant_cols(
    gage_ids[:671], var_lst=['low_prec_freq']
).ravel()
low_prec_dur = camels.read_constant_cols(
    gage_ids[:671], var_lst=['low_prec_dur']
).ravel()
low_prec_timing = camels.read_constant_cols(
    gage_ids[:671], var_lst=['low_prec_timing']
).ravel()
```

```python
climate_data = {
    
    'p_mean':p_mean,
    'pet_mean': pet_mean,
    'p_seasonality':p_seasonality ,
    'frac_snow':frac_snow,
    'aridity': aridity,
    'high_prec_freq':high_prec_freq, 
    'high_prec_dur':high_prec_dur,
    'high_prec_timing':high_prec_timing,
    'low_prec_freq':low_prec_freq,
    'low_prec_dur':low_prec_dur , 
    'low_prec_timing':low_prec_timing
}

climateData = pd.DataFrame(climate_data, columns=['p_mean', 'pet_mean' ,'p_seasonality' ,'frac_snow',
'aridity' ,'high_prec_freq', 'high_prec_dur', 'high_prec_timing',
'low_prec_freq', 'low_prec_dur', 'low_prec_timing'])

print(climateData)

climate_matrix = climateData.corr();
dataplot = sb.heatmap(climate_matrix, cmap="YlGnBu", annot=True)
mp.show()
```

```python
# vegetation Chartacterstics
 #'frac_forest' 'lai_max' 'lai_diff' 'gvf_max' 'gvf_diff'
 #'dom_land_cover_frac' 'dom_land_cover' 'root_depth_50' 'root_depth_99'

frac_forest = camels.read_constant_cols(
    gage_ids[:671], var_lst=['frac_forest']
).ravel()
lai_max = camels.read_constant_cols(
    gage_ids[:671], var_lst=['lai_max']
).ravel()
lai_diff = camels.read_constant_cols(
    gage_ids[:671], var_lst=['lai_diff']
).ravel()
gvf_max = camels.read_constant_cols(
    gage_ids[:671], var_lst=['gvf_max']
).ravel()
gvf_diff = camels.read_constant_cols(
    gage_ids[:671], var_lst=['gvf_diff']
).ravel()
dom_land_cover_frac = camels.read_constant_cols(
    gage_ids[:671], var_lst=['dom_land_cover_frac']
).ravel()
dom_land_cover = camels.read_constant_cols(
    gage_ids[:671], var_lst=['dom_land_cover']
).ravel()
root_depth_50 = camels.read_constant_cols(
    gage_ids[:671], var_lst=['root_depth_50']
).ravel()
root_depth_99 = camels.read_constant_cols(
    gage_ids[:671], var_lst=['root_depth_99']
).ravel()
```

```python


vege_data = {
    
    'frac_forest':frac_forest,
    'lai_max':lai_max,
    'lai_diff':lai_diff,
    'gvf_max':gvf_max,
    'gvf_diff':gvf_diff,
    'dom_land_cover_frac':dom_land_cover_frac,
    'dom_land_cover':dom_land_cover ,
    'root_depth_50':root_depth_50,
    'root_depth_99':root_depth_99
}

vegeData = pd.DataFrame(vege_data,columns=['frac_forest','lai_max','lai_diff','gvf_max','gvf_diff',
 'dom_land_cover_frac', 'dom_land_cover' ,'root_depth_50','root_depth_99'])

print(vegeData)

vege_matrix = vegeData.corr()
dataplot = sb.heatmap(vege_matrix, cmap="YlGnBu", annot=True)
mp.show()

```

```python
# soil characterstics
#'soil_depth_pelletier' 'soil_depth_statsgo' 'soil_porosity'
 #'soil_conductivity' 'max_water_content' 'sand_frac' 'silt_frac'
 #'clay_frac' 'water_frac' 'organic_frac' 'other_frac'

soil_depth_pelletier = camels.read_constant_cols(
    gage_ids[:671], var_lst=['soil_depth_pelletier']
).ravel()
soil_depth_statsgo = camels.read_constant_cols(
    gage_ids[:671], var_lst=['soil_depth_statsgo']
).ravel()
soil_porosity = camels.read_constant_cols(
    gage_ids[:671], var_lst=['soil_porosity']
).ravel()
soil_conductivity = camels.read_constant_cols(
    gage_ids[:671], var_lst=['soil_conductivity']
).ravel()
max_water_content = camels.read_constant_cols(
    gage_ids[:671], var_lst=['max_water_content']
).ravel()
sand_frac = camels.read_constant_cols(
    gage_ids[:671], var_lst=['sand_frac']
).ravel()
silt_frac = camels.read_constant_cols(
    gage_ids[:671], var_lst=['silt_frac']
).ravel()
clay_frac = camels.read_constant_cols(
    gage_ids[:671], var_lst=['clay_frac']
).ravel()
water_frac = camels.read_constant_cols(
    gage_ids[:671], var_lst=['water_frac']
).ravel()
organic_frac = camels.read_constant_cols(
    gage_ids[:671], var_lst=['organic_frac']
).ravel()
other_frac = camels.read_constant_cols(
    gage_ids[:671], var_lst=['other_frac']
).ravel()
```

```python
soil_data = {
    
    'soil_depth_pelletier':soil_depth_pelletier ,
    'soil_depth_statsgo':soil_depth_statsgo,
    'soil_porosity':soil_porosity,
    'soil_conductivity':soil_conductivity ,
    'max_water_content':max_water_content, 
    'sand_frac':sand_frac, 
    'silt_frac':silt_frac,
    'clay_frac':clay_frac,
    'water_frac':water_frac,
    'organic_frac':organic_frac,
    'other_frac':other_frac
}

soilData = pd.DataFrame(soil_data,columns=['soil_depth_pelletier','soil_depth_statsgo','soil_porosity',
'soil_conductivity' ,'max_water_content', 'sand_frac', 'silt_frac',
'clay_frac','water_frac','organic_frac','other_frac'])

print(soilData)

soil_matrix = soilData.corr()
dataplot = sns.heatmap(soil_matrix,cmap="YlGnBu", annot=True)
mp.show()
```

```python
# 'geol_1st_class'
 #'glim_1st_class_frac' 'geol_2nd_class' 'glim_2nd_class_frac'
# 'carbonate_rocks_frac' 'geol_porostiy' 'geol_permeability']
# geological Characterstics
geol_1st_class = camels.read_constant_cols(
    gage_ids[:671], var_lst=['geol_1st_class']
).ravel()
glim_1st_class_frac = camels.read_constant_cols(
    gage_ids[:671], var_lst=['glim_1st_class_frac']
).ravel()
geol_2nd_class = camels.read_constant_cols(
     gage_ids[:671], var_lst=['geol_2nd_class']
).ravel()
glim_2nd_class_frac = camels.read_constant_cols(
     gage_ids[:671], var_lst=['glim_2nd_class_frac']
).ravel()
carbonate_rocks_frac = camels.read_constant_cols(
     gage_ids[:671], var_lst=['carbonate_rocks_frac']
).ravel()
geol_porostiy = camels.read_constant_cols(
     gage_ids[:671], var_lst=['geol_porostiy']
).ravel()
geol_permeability = camels.read_constant_cols(
     gage_ids[:671], var_lst=['geol_permeability']
).ravel()
```

```python
# displaying heatmap
# geological Characterstics

data = {
    'geol_1st_class': geol_1st_class,
    'glim_1st_class_frac': glim_1st_class_frac,
    'geol_2nd_class': geol_2nd_class,
    'glim_2nd_class_frac': geol_2nd_class,
    'carbonate_rocks_frac': carbonate_rocks_frac,
    'geol_porostiy':geol_porostiy, 
    'geol_permeability': geol_permeability

}
 
# form dataframe
dataframe = pd.DataFrame(data, columns=['geol_1st_class',
    'glim_1st_class_frac',
    'geol_2nd_class',
    #'glim_2nd_class_frac',
    'carbonate_rocks_frac',
    'geol_porostiy', 
    'geol_permeability'])
print("Dataframe is : ")
print(dataframe)

# form correlation matrix
matrix = dataframe.corr()
print("Correlation matrix is : ")
print(matrix)
dataplot = sb.heatmap(matrix, cmap="YlGnBu", annot=True)
  
# displaying heatmap
mp.show()
```
